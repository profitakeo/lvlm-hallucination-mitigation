```python
import json
import os
from PIL import Image
import requests
from io import BytesIO
import torch

# =============================================================================
# Baseline POPE Evaluation
# =============================================================================

def download_pope_questions():
    """Download POPE questions for COCO validation set"""

    # Official POPE evaluation questions
    pope_urls = {
        "random": "https://github.com/AoiDragon/POPE/raw/e3e39262c85a6a83f26cf5094022a782cb0df58d/output/coco/coco_pope_random.json",
        "popular": "https://github.com/AoiDragon/POPE/raw/e3e39262c85a6a83f26cf5094022a782cb0df58d/output/coco/coco_pope_popular.json",
        "adversarial": "https://github.com/AoiDragon/POPE/raw/e3e39262c85a6a83f26cf5094022a782cb0df58d/output/coco/coco_pope_adversarial.json"
    }
    # Random objects that DON'T exist in the image
    # Common/popular objects that DON'T exist in this specific image
    # Objects that frequently appear WITH objects in the image, but aren't actually there

    pope_data = {}

    for pope_type, url in pope_urls.items():
        print(f"Downloading POPE {pope_type} questions...")
        response = requests.get(url)

        # Parse JSONL format (each line is a separate JSON object)
        pope_data[pope_type] = []
        lines = response.text.strip().split('\n')

        for line in lines:
            if line.strip():
                try:
                    pope_data[pope_type].append(json.loads(line))
                except json.JSONDecodeError:
                    print(f"Error parsing line: {line[:50]}...")
                    continue

        print(f"Downloaded {len(pope_data[pope_type])} questions for {pope_type}")

    return pope_data

def load_coco_image(image_id, coco_images_dir="val2017"):
    """Load COCO image by ID"""

    # Handle different formats
    if isinstance(image_id, str) and "COCO_val" in image_id:
        # Extract ID from "COCO_val2014_000000310196.jpg" -> 310196
        image_id = int(image_id.split('_')[-1].split('.')[0])

    # COCO 2017 filename format: 000000123456.jpg (12 digits)
    image_filename = f"{image_id:012d}.jpg"
    image_path = os.path.join(coco_images_dir, image_filename)

    if os.path.exists(image_path):
        return Image.open(image_path).convert("RGB")
    else:
        print(f"Image {image_filename} not found in {coco_images_dir}")
        return None

def evaluate_pope_question(image, question, model, processor):
    """Evaluate single POPE question"""

    # prompt = f"PROMPT: <image>\n{question} {prompt}\nRESPONSE:"

    prompt = f"""USER: <image>
Question: {question}
Describe what you see in the image, then confirm your final answer.
RESPONSE: """

    # Process inputs
    inputs = processor(text=prompt, images=image, return_tensors="pt") # pt: PyTorch tensors

    # print(inputs.keys())
    # input_ids: [1, 15, 25, 42, 50, 67, ...] (prompt text as token IDs)
    # attention_mask: [1, 1, 1, 1, 1, ...] (which prompt tokens to attend to, or discard e.g. special tokens)
    # pixel_values: [[[0.2, 0.1, ...], ...]] (input image as normalized RGB arrays)

    # Move tensors to GPU if available, for faster processing (else, leave strings, metadata etc as is)
    if torch.cuda.is_available():
        inputs = {k: v.to("cuda") if isinstance(v, torch.Tensor) else v for k, v in inputs.items()}

    # Generate model response
    with torch.no_grad():                 # disable gradient computation (not required for inference)
        generated_ids = model.generate(
            **inputs,                     # unpacks input dictionary
            max_new_tokens=50,
            min_new_tokens=20,
            do_sample=False               # deterministic generation
        )

    # Decode model response
    result = processor.decode(generated_ids[0], skip_special_tokens=True)

    # Get only the generated part (after RESPONSE:)
    if "RESPONSE:" in result:
        response = result.split("RESPONSE:")[-1].strip()
    else:
        response = result.strip()

    response_lean = response.lower()#.replace(" ", "")

    if "yes" in response_lean:
        final_answer = "yes"
    elif "no" in response_lean:
        final_answer = "no"
    else:
        final_answer = "unsure"

    print(f"Full generated text: {result}")
    print(f"Extracted answer: {response}")
    print(f"Final extracted answer: {final_answer}")

    return final_answer
```

```python
import numpy as np
import random
import torch.nn.functional as F
from transformers import LogitsProcessor

# =============================================================================
# VCD (Visual Contrastive Decoding) Implementation
# =============================================================================

'''
* Creates noisy version of the image using random Gaussian noise
* Compares logits between clean and noisy image, using: (1 + α) * logit_clean - α * logit_distorted
* If token logit is higher for noisy image, model is likely hallucincating (falls back on language prior reliance)
* If token logit is higher for clean image, model is likely responding to visual correctly
* Alpha controls strength of penalty and boosting
* Aims to reduce false positives

Example: Is there a motorcycle?
Actual answer: No (there's a car)

Clean img: motorcycle logit = 1.2
Noisy img: motorcycle logit = 2.8 (model falls back on learnt co-occurrence betwen car and motorcycle)

Final output logit (after penalty): (1 + 1.0) * 1.2 - 1.0 * 2.8 = 2.4 - 2.8 = -0.4

'''

def add_gaussian_noise(image, noise_mean, noise_std, counter):
    """Add random Gaussian noise to original image"""

    # Convert PIL Image to numpy array
    img_array = np.array(image).astype(np.float32) / 255.0

    # Add random Gaussian noise; ensure pixel values are [0,1]
    noise = np.random.normal(noise_mean, noise_std, img_array.shape)
    noisy_img = np.clip(img_array + noise, 0, 1)

    # Print sample stats for debugging
    if counter==0:
      print(f"\n[CLEAN IMG STATS]")
      print(f"  - Shape: {img_array.shape}")
      print(f"  - Min pixel value: {img_array.min():.3f}")
      print(f"  - Max pixel value: {img_array.max():.3f}")
      print(f"  - Mean pixel value: {img_array.mean():.3f}")
      print(f"\n[NOISE STATS]")
      print(f"  - Noise mean: {noise.mean():.3f}")
      print(f"  - Noise std: {noise.std():.3f}")
      print(f"  - Noise min: {noise.min():.3f}")
      print(f"  - Noise max: {noise.max():.3f}")
      print(f"\n[NOISY IMG STATS]:")
      print(f"  - Min pixel value: {noisy_img.min():.3f}")
      print(f"  - Max pixel value: {noisy_img.max():.3f}")
      print(f"  - Mean pixel value: {noisy_img.mean():.3f}")

    # Convert numpy array back to PIL Image
    noisy_img = (noisy_img * 255).astype(np.uint8)

    return Image.fromarray(noisy_img)

class VCDLogitsProcessor(LogitsProcessor):
    def __init__(self, model, input_distorted, alpha, counter, processor=None):
        self.model = model
        self.input_distorted = input_distorted
        self.alpha = alpha
        self.counter = counter

        self.processor = processor
        self.step_count = 0  # Track generation step for each example

    def __call__(self, input_ids, scores):
        """
        Args:
            input_ids: Current sequence being generated [batch_size, seq_len]
            scores:    Current logits from clean image [batch_size, vocab_size]

        Returns:
            Modified scores after VCD adjustment (i.e. penalty or boosting)
        """

        self.step_count += 1

        with torch.no_grad():
            # Update distorted inputs with current input sequence
            self.input_distorted["input_ids"] = input_ids

            # Get logits from distorted image
            output_distorted = self.model(**self.input_distorted)
            logits_distorted = output_distorted.logits[:, -1, :]  # [batch_size, last_position ie next token, vocab_size]

            # Apply VCD formula
            vcd_scores = (1 + self.alpha) * scores - self.alpha * logits_distorted

            if self.counter==0:
              # Track generation for each example
              if self.processor: # and self.step_count <= 10:
                  current_text = self.processor.decode(input_ids[0], skip_special_tokens=True)
                  print(f"\n[VCD Step {self.step_count}]")

                  # Get top 10 tokens
                  top_k = 10
                  top_values, top_indices = torch.topk(scores[0], top_k)

                  print(f"\n  Top 10 tokens:")
                  print(f"  {'Token':<20} {'Clean':<8} {'Noisy':<8} {'VCD':<8} {'Change':<8}")
                  print("  " + "-"*60)

                  for i in range(top_k):
                      idx = top_indices[i].item()
                      token = self.processor.decode([idx], skip_special_tokens=True)
                      if not token:  # Handle special tokens
                          token = f"[{idx}]"
                      clean_logit = scores[0, idx].item()
                      noisy_logit = logits_distorted[0, idx].item()
                      vcd_score = vcd_scores[0, idx].item()
                      change = vcd_score - clean_logit

                      print(f"  {token:<20} {clean_logit:<8.3f} {noisy_logit:<8.3f} {vcd_score:<8.3f} {change:<8.3f}")

            return vcd_scores

def evaluate_pope_question_vcd(image, question, model, processor, counter, alpha, noise_mean, noise_std):
    """
    Visual Contrastive Decoding generation

    Args:
        model: LLaVA model
        processor: LLaVA processor
        image: PIL Image
        question: Question text
        counter: question index of run
        alpha: VCD strength parameter
        noise_mean: Gaussian noise mean for distorted image
        noise_std: Gaussian noise std/level for distorted image
    """

    # prompt = f"USER: <image>\n{question} {prompt}\nASSISTANT:"

    prompt = f"""USER: <image>
Question: {question}
Describe what you see in the image, then confirm your final answer.
RESPONSE: """

    # Original image input
    input_clean = processor(text=prompt, images=image, return_tensors="pt") # pt: PyTorch tensors

    # Distorted image input
    img_distorted = add_gaussian_noise(image, noise_mean, noise_std, counter)
    input_distorted = processor(text=prompt, images=img_distorted, return_tensors="pt")

    if counter==0:
      image.save(f"vcd_clean_img_alpha_{alpha}_noise_{noise_std}.jpg")
      img_distorted.save(f"vcd_noisy_img_alpha_{alpha}_noise_{noise_std}.jpg")

    # Move tensors to GPU if available, for faster processing (else, leave string, metadata etc as is)
    if torch.cuda.is_available():
        input_clean = {k: v.to("cuda") if isinstance(v, torch.Tensor) else v for k, v in input_clean.items()}
        input_distorted = {k: v.to("cuda") if isinstance(v, torch.Tensor) else v for k, v in input_distorted.items()}

    # Create VCD processor to obtain adjusted scores
    vcd_processor = VCDLogitsProcessor(model, input_distorted, alpha, counter, processor)

    # Generate model response
    with torch.no_grad():                      # disable gradient computation
      generated_ids = model.generate(
          **input_clean,                       # unpacks clean input dictionary
          max_new_tokens=50,
          min_new_tokens=20,
          logits_processor=[vcd_processor],    # apply VCD adjustment
          do_sample=False,                     # deterministic generation
          # temperature=0.7
      )

    # Decode model response
    result = processor.decode(generated_ids[0], skip_special_tokens=True)

    # Get only the generated part (after RESPONSE:)
    if "RESPONSE:" in result:
        response = result.split("RESPONSE:")[-1].strip()
    else:
        response = result.strip()

    response_lean = response.lower()#.replace(" ", "")

    if "yes" in response_lean:
        final_answer = "yes"
    elif "no" in response_lean:
        final_answer = "no"
    else:
        final_answer = "unsure"

    print(f"Full generated text: {result}")
    print(f"Extracted answer: {response}")
    print(f"Final extracted answer: {final_answer}")

    return final_answer
```

```python
def run_pope_evaluation(model, processor, pope_data, pope_type, method, num_questions, alpha, noise_mean, noise_std):
    """Run a simple POPE evaluation"""

    print(f"Running POPE evaluation on {num_questions} questions...")

    results = []
    j=0

    questions = pope_data[pope_type]

    for i, item in enumerate(questions):

        # Handle image filename format
        if isinstance(item['image'], str) and "COCO_val" in item['image']:
            # Extract ID from "COCO_val2014_000000310196.jpg" -> 310196
            item['image'] = int(item['image'].split('_')[-1].split('.')[0])

        print(f"\nQuestion {j+1}/{num_questions}: Processing image {item['image']:012d}.jpg")

        # Load image
        image = load_coco_image(item["image"])

        if image is None:
            print(f"Skipping question {i+1} - image not found")
            continue

        # Get model's answer
        if method == 'baseline':
          predicted_answer = evaluate_pope_question(image, item["text"], model, processor)
        elif method == 'vcd':
          predicted_answer = evaluate_pope_question_vcd(image, item["text"], model, processor, j, alpha, noise_mean, noise_std)

        # Store result
        result = {
            "image_id": item["image"],
            "question": item["text"],
            "ground_truth": item["label"],
            "predicted": predicted_answer,
            "correct": predicted_answer == item["label"]
        }
        results.append(result)

        j+=1

        if j==num_questions:
          break

    return results

def calculate_pope_metrics(results):
    """Calculate POPE evaluation metrics"""

    # Count outcomes
    true_positives = sum(1 for r in results if r["ground_truth"] == "yes" and r["predicted"] == "yes")
    false_positives = sum(1 for r in results if r["ground_truth"] == "no" and r["predicted"] == "yes")
    true_negatives = sum(1 for r in results if r["ground_truth"] == "no" and r["predicted"] == "no")
    false_negatives = sum(1 for r in results if r["ground_truth"] == "yes" and r["predicted"] == "no")

    # Calculate metrics
    accuracy = (true_positives + true_negatives) / len(results)
    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
    f1_score = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

    yes_ratio = sum(1 for r in results if r["predicted"] == "yes") / len(results)

    return {
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1_score": f1_score,
        "yes_ratio": yes_ratio,
        "true_positives": true_positives,
        "false_positives": false_positives,
        "true_negatives": true_negatives,
        "false_negatives": false_negatives
    }

# Main evaluation function
def main_pope_evaluation(model, processor, pope_type, method, num_questions, alpha=1.0, noise_mean=0, noise_std=0.1):
    """Main function to run POPE evaluation"""

    print(f"Starting {pope_type} POPE Evaluation on ...")

    # Step 1: Download POPE questions
    pope_data = download_pope_questions()

    # Step 2: Load COCO images and run evaluation
    results = run_pope_evaluation(model, processor, pope_data, pope_type, method, num_questions, alpha, noise_mean, noise_std)

    # Step 3: Calculate metrics
    metrics = calculate_pope_metrics(results)

    # Step 4: Print results
    print("\n" + "="*50)
    print("POPE EVALUATION RESULTS")
    print("="*50)
    print(f"Accuracy: {metrics['accuracy']:.4f}")
    print(f"Precision: {metrics['precision']:.4f}")
    print(f"Recall: {metrics['recall']:.4f}")
    print(f"F1 Score: {metrics['f1_score']:.4f}")
    print(f"Yes Ratio: {metrics['yes_ratio']:.4f}")
    print(f"True Positives: {metrics['true_positives']}")
    print(f"False Positives: {metrics['false_positives']}")
    print(f"True Negatives: {metrics['true_negatives']}")
    print(f"False Negatives: {metrics['false_negatives']}")

    return results, metrics
```

```python
# Run baseline evaluation on POPE

import sys
from contextlib import redirect_stdout

pope_type="popular"
method="baseline"
num_questions=445

with open(f'{method}_{pope_type}_{num_questions}qs.txt', 'w') as f:
    with redirect_stdout(f):
        results, metrics = main_pope_evaluation(model=model, processor=processor, pope_type=pope_type, method=method, num_questions=num_questions)

from google.colab import files
files.download(f'{method}_{pope_type}_{num_questions}qs.txt')
```
